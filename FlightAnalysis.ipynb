#### So let's consider the Date_of_Journey column it is of object datatype we need to convert it into date time format and create 3 derived features from it that is day,month and year separately

So let's start feature engineering process

```python
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

### Or you can also use
1. final_df['Date'] = final_df['Date_of_Journey'].apply(lamda x:x.split('/')[0])
2. final_df['Month'] = final_df['Date_of_Journey'].apply(lamda x:x.split('/')[1])
3. final_df['Year'] = final_df['Date_of_Journey'].apply(lamda x:x.split('/')[2])

```python
final_df.info()
```
